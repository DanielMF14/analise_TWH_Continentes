# Já analisamos o consumo de TWH por continentes, agora vamos ver o consumo de TWH em alguns países.

```python
# Importamos as bibliotecas necessárias:
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy as scp
```

```python
# Extraimos os nossos dados:
df_Consumo_Paises = pd.read_csv("Country_Consumption_TWH.xls", index_col='Year')
pd.set_option("display.max_columns", None)
df_Consumo_Paises
```

* Vamos eliminar alguns valores faltantes: 

```python
df_Consumo_Paises = df_Consumo_Paises.dropna()
```

```python
df_Consumo_Paises
```

* Vamos usar a função describe para termos um breve resumo do nosso DataFrame:

```python
df_Consumo_Paises.describe().round(3)
```

* Vamos analisar como foi o consumo de TWH no Brasil ao longo de 30 anos:

```python
df_Consumo_Brasil = df_Consumo_Paises['Brazil']
df_Consumo_Brasil
```

* Agora vamos ver esses dados plotados em um gráfico de linhas: 

```python
fig, ax = plt.subplots(figsize=(18,6))

plt.suptitle("Consumo de terawatt(em milhares) no Brasil ao longo de 30 anos.", fontsize=20, weight='semibold')

ax.plot(df_Consumo_Brasil, label='Consumo anual', marker='o')

ax.set_xlabel("Anos", fontsize=14)
ax.set_ylabel("Consumo", fontsize=14)

ax.set_xticks(df_Consumo_Brasil.index)
ax.set_xticklabels(df_Consumo_Brasil.index, rotation=45)

ax.legend()
plt.show()
```

* Vamos comparar o consumo de TWH no Brasil, ao longo dos 30 anos, em relação a outros países do continente americano:

```python
# Primeiro vamos filtrar os países de nosso interesse:
soma = df_Consumo_Paises[['Brazil', 'United States', 'Argentina', 'Chile', 'Colombia', 'Mexico', 'Venezuela', 'Canada']].sum().round(3).sort_values()
```

```python
# Vamos plotar um gráfico para vermos a distruição de consumo:

plt.figure(figsize=(18,6))
plt.bar(soma.index, soma.values, color='skyblue')

plt.title("Consumo total dos países do continente americano ao longo de 30 anos", fontsize=20, weight="bold")
plt.xlabel("Países", fontsize=15)
plt.ylabel("Consumo Total", fontsize=15)

for i, valor in enumerate(soma.values):
    plt.text(i, valor + 2, str(valor), ha='center', fontsize=12)
plt.show()
```

# Vamos ver o consumo de Terawatts/Hora ao longo de 30 anos nos países do nosso DataFrame. 

```python
# loop para criar um gráfico para cada país:
for pais in df_Consumo_Paises.columns:
    plt.figure(figsize=(12,6))
    plt.plot(df_Consumo_Paises.index, df_Consumo_Paises[pais], marker='o', label='Evolução do consumo')
    plt.title(f"Consumo de TWH no país: {pais}", fontsize=18, weight='bold')
    plt.xlabel('Anos', fontsize=14)
    plt.ylabel('Consumo(TWH)',fontsize=14)
    plt.legend(fontsize=12)
    plt.grid(True)
    plt.show()
```

# Vamos criar um heatmap para ver a intensidade do consumo no Brasil.

```python
# Vamos copiar um novo df para ultilizarmos no nosso mapa de calor:
df_Consumo_Brasil1 = df_Consumo_Brasil.copy().reset_index()
```

```python
# Vamos renomear as colunas do nosso df:
df_Consumo_Brasil1.columns = ['Ano','Consumo TWH']
```

```python
df_Consumo_Brasil1
```

```python
# Plotando heatmap:
plt.figure(figsize=(9,9))
sns.heatmap(df_Consumo_Brasil1, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Mapa de calor do consumo de TWH no Brasil ao longo de 30 anos.')
plt.show()
```
