# Vamos fazer uma an√°lise sobre o consumo de twh por continente.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy as scp
```

```python
# Criamos um DataFrame que recebe a nossa base de dados e passamos a coluna de anos como o nosso √≠ndice:
df_Continente = pd.read_csv("Continent_Consumption_TWH.xls", index_col="Year")
df_Continente
```

* Vamos fazer uma an√°lise do consumo(TWH) por continente. Mas antes, vamos ver o consumo total por ano.

```python
# Vamos criar um novo DF que cont√©m os dados total de consumo:
df_Consumo_World = df_Continente['World']
df_Consumo_World
```

* Vamos plotar um gr√°fico de linha para ver nossas s√©ries temporais:

```python
# Criar uma figura e √°rea de plotagem:
fig, ax = plt.subplots(figsize=(16,8)) 

# Adicionando titulo √† figura:
plt.suptitle("Consumo de Terawatt-hora (TWh) no mundo ao longo de 30 anos.", fontsize=20, weight='bold')

# Plotar a s√©rie temporal(Gr√°fico de linhas):
ax.plot(df_Consumo_World, label='Consumo anual')

# D√° titulo aos eixos:
ax.set_xlabel("Anos", fontsize=14)
ax.set_ylabel("Consumo(TWH)", fontsize=14)

# Adicionar todos os anos no eixo X
ax.set_xticks(df_Consumo_World.index)  # Assumindo que o √≠ndice do DataFrame cont√©m os anos
ax.set_xticklabels(df_Consumo_World.index, rotation=45)  # Gira os anos para melhor visualiza√ß√£o

# Adicionando a legenda:
ax.legend()
#Plotando:
plt.show()
```

* Agora vamos ver o consumo de TWH na Europa.

```python
# Novamente vamos filtrar somente um vari√°vel, no caso, o continente europeu:
df_Consumo_Europa = df_Continente['Europe']
df_Consumo_Europa
```

```python
fig, ax = plt.subplots(figsize=(16,8))

plt.suptitle("Consumo de Terawatt-hora (TWh) na Europa ao longo de 30 anos.", fontsize=20, weight='semibold')

ax.plot(df_Consumo_Europa, label='Consumo anual europeu')

ax.set_xlabel('Anos', fontsize=14, labelpad=20)
ax.set_ylabel('Consumo por ano', fontsize=14, labelpad=20)

# Adicionar todos os anos no eixo X
ax.set_xticks(df_Consumo_Europa.index)  # Assumindo que o √≠ndice do DataFrame cont√©m os anos
ax.set_xticklabels(df_Consumo_Europa.index, rotation=45)  # Gira os anos para melhor visualiza√ß√£o

ax.legend()
plt.show()
```

* Vamos ver a m√©dia m√≥vel do consumo anual de TWH. 

```python
# Calculando as m√©dias m√≥veis:
df_rolling_12m = df_Consumo_Europa.rolling(window=12).mean()
df_rolling_6m = df_Consumo_Europa.rolling(window=6).mean()
```

```python
fig, ax = plt.subplots(figsize=(16,8))

plt.suptitle("Consumo de Terawatt-hora (TWh) na Europa ao longo de 30 anos.", fontsize=20, weight='semibold')

ax.plot(df_Consumo_Europa.index, df_Consumo_Europa, label='Consumo anual europeu')
ax.plot(df_Consumo_Europa.index,df_rolling_6m, label='M√©dia m√≥vel (6 meses)')
ax.plot(df_Consumo_Europa.index,df_rolling_12m, label='M√©dia m√≥vel (12 meses)')

# Adicionar todos os anos no eixo X
ax.set_xticks(df_Consumo_World.index)  # Assumindo que o √≠ndice do DataFrame cont√©m os anos
ax.set_xticklabels(df_Consumo_World.index, rotation=45)  # Gira os anos para melhor visualiza√ß√£o

ax.set_xlabel('Anos', fontsize=14, labelpad=20)
ax.set_ylabel('Consumo por ano', fontsize=14, labelpad=20)

ax.legend()
plt.show()
```

* Vamos ver o consumo de TWH na Am√©rica do Norte. 

```python
df_Consumo_AmericaNorte = df_Continente['North America']
df_Consumo_AmericaNorte
```

```python
# Para automatizar o processo de cria√ß√£o de gr√°ficos vamos criar uma fun√ß√£o para plotar o consumo de TWH p/continente: 
def plot_time_series(data, y_col, title, xlabel, ylabel, figsize=(18, 6), label_line=None):
    """
    Fun√ß√£o para plotar gr√°ficos de s√©ries temporais.
    
    Par√¢metros:
    - data (pd.DataFrame): DataFrame contendo os dados.
    - y_col (str): Nome da coluna a ser plotada no eixo Y.
    - title (str): T√≠tulo do gr√°fico.
    - xlabel (str): R√≥tulo do eixo X.
    - ylabel (str): R√≥tulo do eixo Y.
    - figsize (tuple): Tamanho da figura (largura, altura).
    - label_line (str): R√≥tulo da linha no gr√°fico (para legenda).
    
    Retorno:
    - Nenhum. Exibe o gr√°fico.
    """
    # Cria figura e √°rea de plotagem
    fig, ax = plt.subplots(figsize=figsize)

    # Adiciona t√≠tulo √† figura
    plt.suptitle(title, fontsize=20, weight='bold')

    # Plota s√©rie temporal
    ax.plot(data, label=label_line if label_line else y_col)

    # Define os t√≠tulos dos eixos
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)

    # Configura o eixo X para mostrar todos os anos
    ax.set_xticks(data.index)  # Define os ticks como todos os valores do √≠ndice
    ax.set_xticklabels(data.index, rotation=45)  # Rotaciona para legibilidade

    # Adiciona legenda
    if label_line or y_col:
        ax.legend()

    # Mostra o gr√°fico
    plt.show()
```

```python
plot_time_series(df_Consumo_AmericaNorte,"Consumo de TWH", "Consumo de TWH anual Norte Americano", "Ano", "Consumo", (18,8), 'Varia√ß√£o de consumo')
```

* Vamos ver a m√©dia m√≥vel na Am√©rica do Norte.

```python
df_rolling_AN_6m = df_Consumo_AmericaNorte.rolling(window=6).mean()
df_rolling_AN_12m = df_Consumo_AmericaNorte.rolling(window=12).mean()
```

```python
fig, ax = plt.subplots(figsize=(16,8))

plt.suptitle("Consumo de Terawatt-hora (TWh) na Am√©rica do Norte ao longo de 30 anos.", fontsize=20, weight='semibold')

ax.plot(df_Consumo_AmericaNorte, label='Consumo anual norte americano.')
ax.plot(df_rolling_AN_6m, label='M√©dia m√≥vel (6 meses)')
ax.plot(df_rolling_AN_12m, label='M√©dia m√≥vel (12 meses)')

# Adicionar todos os anos no eixo X
ax.set_xticks(df_Consumo_AmericaNorte.index)  # Assumindo que o √≠ndice do DataFrame cont√©m os anos
ax.set_xticklabels(df_Consumo_AmericaNorte.index, rotation=45)  # Gira os anos para melhor visualiza√ß√£o

ax.set_xlabel('Anos', fontsize=14, labelpad=20)
ax.set_ylabel('Consumo por ano', fontsize=14, labelpad=20)

ax.legend()
plt.show()
```
