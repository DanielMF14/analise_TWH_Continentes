# Vamos fazer uma análise sobre o consumo de twh por continente.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy as scp
```

```python
# Criamos um DataFrame que recebe a nossa base de dados e passamos a coluna de anos como o nosso índice:
df_Continente = pd.read_csv("Continent_Consumption_TWH.xls", index_col="Year")
df_Continente
```

# Vamos fazer uma análise do consumo(TWH) por continente. Mas antes, vamos ver o consumo total por ano.

```python
# Vamos criar um novo DF que contém os dados total de consumo:
df_Consumo_World = df_Continente['World']
df_Consumo_World
```

* Vamos plotar um gráfico de linha para ver nossas séries temporais:

```python
# Criar uma figura e área de plotagem:
fig, ax = plt.subplots(figsize=(16,8)) 

# Adicionando titulo à figura:
plt.suptitle("Consumo de Terawatt-hora (TWh) no mundo ao longo de 30 anos.", fontsize=20, weight='bold')

# Plotar a série temporal(Gráfico de linhas):
ax.plot(df_Consumo_World, label='Consumo anual', marker='o')

# Dá titulo aos eixos:
ax.set_xlabel("Anos", fontsize=14)
ax.set_ylabel("Consumo(TWH)", fontsize=14)

# Adicionar todos os anos no eixo X
ax.set_xticks(df_Consumo_World.index)  # Assumindo que o índice do DataFrame contém os anos
ax.set_xticklabels(df_Consumo_World.index, rotation=45)  # Gira os anos para melhor visualização

# Adicionando a legenda:
ax.legend()
#Plotando:
plt.show()
```

# Vamos ver o consumo de TWH na América do Norte. 

```python
df_Consumo_AmericaNorte = df_Continente['North America']
df_Consumo_AmericaNorte
```

```python
# Para automatizar o processo de criação de gráficos vamos criar uma função para plotar o consumo de TWH p/continente: 
def plot_time_series(data, y_col, title, xlabel, ylabel, figsize=(18, 6), label_line=None):
    """
    Função para plotar gráficos de séries temporais.
    
    Parâmetros:
    - data (pd.DataFrame): DataFrame contendo os dados.
    - y_col (str): Nome da coluna a ser plotada no eixo Y.
    - title (str): Título do gráfico.
    - xlabel (str): Rótulo do eixo X.
    - ylabel (str): Rótulo do eixo Y.
    - figsize (tuple): Tamanho da figura (largura, altura).
    - label_line (str): Rótulo da linha no gráfico (para legenda).
    
    Retorno:
    - Nenhum. Exibe o gráfico.
    """
    # Cria figura e área de plotagem
    fig, ax = plt.subplots(figsize=figsize)

    # Adiciona título à figura
    plt.suptitle(title, fontsize=20, weight='bold')

    # Plota série temporal
    ax.plot(data, label=label_line if label_line else y_col, marker='o')

    # Define os títulos dos eixos
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)

    # Configura o eixo X para mostrar todos os anos
    ax.set_xticks(data.index)  # Define os ticks como todos os valores do índice
    ax.set_xticklabels(data.index, rotation=45)  # Rotaciona para legibilidade

    # Adiciona legenda
    if label_line or y_col:
        ax.legend()

    # Mostra o gráfico
    plt.show()
```

```python
plot_time_series(df_Consumo_AmericaNorte,"Consumo de TWH", "Consumo de TWH anual Norte Americano", "Ano", "Consumo", (18,8), 'Variação de consumo')
```

* Vamos ver a média móvel na América do Norte.

```python
df_rolling_AN_6m = df_Consumo_AmericaNorte.rolling(window=6).mean()
df_rolling_AN_12m = df_Consumo_AmericaNorte.rolling(window=12).mean()
```

```python
fig, ax = plt.subplots(figsize=(16,8))

plt.suptitle("Consumo de Terawatt-hora (TWh) na América do Norte ao longo de 30 anos.", fontsize=20, weight='semibold')

ax.plot(df_Consumo_AmericaNorte, label='Consumo anual norte americano.', marker='o')
ax.plot(df_rolling_AN_6m, label='Média móvel (6 meses)', marker='o')
ax.plot(df_rolling_AN_12m, label='Média móvel (12 meses)', marker='o')

# Adicionar todos os anos no eixo X
ax.set_xticks(df_Consumo_AmericaNorte.index)  # Assumindo que o índice do DataFrame contém os anos
ax.set_xticklabels(df_Consumo_AmericaNorte.index, rotation=45)  # Gira os anos para melhor visualização

ax.set_xlabel('Anos', fontsize=14, labelpad=20)
ax.set_ylabel('Consumo por ano', fontsize=14, labelpad=20)

ax.legend()
plt.show()
```

# Vamos ver o total de consumo de terawalts/hora de cada continente.

```python
# Primeiro vamos somar o consumo total de todas as variáveis:
soma = df_Continente[['Europe', 'North America', 'Latin America', 'Asia', 'Pacific', 'Africa', 'Middle-East']].sum().round(3).sort_values()
soma
```

* Vamos plotar um gráfico para melhor visualização.

```python
# Configurando o gráfico de barras:
plt.figure(figsize=(18,8))
plt.bar(soma.index, soma.values, color='skyblue')

# Adicionando títulos e rótulos:
plt.title("Consumo total por continentes e blocos econômicos ao longo de 30 anos", fontsize=20, weight="bold")
plt.xlabel("Continentes e Blocos econômicos", fontsize=15)
plt.ylabel("Consumo Total", fontsize=15)

# Mostrar os valores no topo de cada barra
for i, valor in enumerate(soma.values):
    plt.text(i, valor + 2, str(valor), ha="center", fontsize=12)

# Plotar gráfico:
plt.show()
```
